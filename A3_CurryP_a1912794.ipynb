```python
# Rubric Dictionary
rainfall = [
{'Sydney' : 50,'Melbourne' : 30,'Adelaide' : 22},
{'Sydney' : 10,'Perth' : 14},
{'Adelaide' : 30,'Melbourne' : 50},
{'Sydney' : 20,'Melbourne' : 30,'Brisbane' :50},
{'Perth' : 20,'Adelaide' : 30,'Brisbane' :50},
{'Melbourne' : 30,'Adelaide' : 22},
{'Sydney' : 50,'Brisbane' : 33,'Adelaide' : 28}]

#check Perth total is 34mm
for p in rainfall:
    if 'Perth' in p:
        print({p["Perth"]})
```

```python
def avg_rainfall_calculator(rainfall):
    #Create empty dictionaries
    rainfall_mm = {}
    rainfall_event = {}
    #loop through rainfall dictionary
    for event in rainfall:
        # iterate through each event
        for city, value in event.items():
            # Error handling (negative integers and characters)
            if not (isinstance(value, (int, float)) and value >= 0):
                print("Invalid rainfall for", city, ":", value, ". Entry ommitted from calculation")
            else:
                # Use get method to add mm value from city key. return value. If nothing return 0.
                rainfall_mm[city] = rainfall_mm.get(city, 0) + value
                # For every instance, add 1 to event key
                rainfall_event[city] = rainfall_event.get(city, 0) + 1

    # Use dictionaries to calculate key:value averages. Round method returns results to 2dp
    averages = {city: round(rainfall_mm[city] / rainfall_event[city], 2) for city in rainfall_mm}    
    # Make it a globally available variable 
    return averages

print("averages= ", averages)
```

# B- Exploring vectorisation

```python
import numpy

# Generate random 2D NumPy array of size (m,n) with random integers ranging from 0 to 99.
Original_array = np.random.randint(0, 100, size=(4, 5))
print("Original array: \n", Original_array)

# Reshape the array to size (m*n, 1) using the reshape() function
Reshaped_array = Original_array.reshape((20,1))
print("Reshaped array: \n", Reshaped_array)

# Compute mean, median & standard deviation of array using NumPy functions.
Reshaped_array_mean = np.mean(Reshaped_array)
Reshaped_array_median = np.median(Reshaped_array)
Reshaped_array_std = np.std(Reshaped_array)
print("Data insights of Reshaped_array: \n", "Mean: ", Reshaped_array_mean, "\n Median: ", Reshaped_array_median, "\n Standard deviation: ", Reshaped_array_std)

# Slice array to get first k elements, where k is total number of elements in array divided by 2
k = Reshaped_array.size // 2
    # Error handling for empty array(s)
if k > 0:
    Sliced_array = Reshaped_array[:k]
    print("Sliced array: \n", Sliced_array)
else:
    print("Oh no... I can't slice an array with 0 elements.")

# Split into equal sized arrays of a random size between 2 and 5
    # generate random integer, save to p
p = np.random.randint(2, 6)
    # calling array_split function splits the array into equal parts (based on value of p)
split_parts = np.array_split(Reshaped_array, p)
print("Split array into", p, "parts: ")
    # iterate through array & print elements
for i, part in enumerate(split_parts):
    print("Array", i,"\n", part)

# Compute the element-wise sum of the split arrays using vectorisation.    
    # 0 axis set to specify summing along a 2D array. If not specified, it sums everything (1D) 
sum_of_splits = np.sum(split_parts, axis=0)
print("Element-wise sum of split arrays: \n", sum_of_splits)
```

```python
# Data cleanning- remove null values
ab_NYC_2019_clean = ab_NYC_2019.dropna()
# Data manipulation- remove columns not needed
ab_NYC_2019_tamed = ab_NYC_2019_clean.drop(columns=['availability_365', 'calculated_host_listings_count', 'last_review', 'latitude', 'longitude', 'minimum_nights', 'name', 'reviews_per_month', 'room_type'])
display(ab_NYC_2019_tamed)
```

```python
# Summary statistics for ‘rental price’ and ‘number of reviews’ 

# pandas describe funcrtion to generate stats summary
price_statistics = ab_NYC_2019_tamed['price'].describe()
review_count_statistics = ab_NYC_2019_tamed['number_of_reviews'].describe()
# concatenate results into a single output
combined_statistics = pd.concat([price_statistics, review_count_statistics], axis=1)
print("Air BnB statistics - NYC 2019 \n")
print(combined_statistics)
```

```python
# Average rental price and reviews by neighbourhood

# groupby() pandas function creates object
hood_data = ab_NYC_2019_tamed.groupby('neighbourhood')
# use agg method on the object to create key as column name and perform a calculation to populate value(mean)) 
average_statistics_by_hood = hood_data.agg({ 'price': 'mean', 'number_of_reviews': 'mean'})
print("AirBnB Average Price and Review by Neighbourhood")
display(average_statistics_by_hood)
```
